```python
dictarg1 = bws["argument1"].to_dict()
dictarg2 = bws["argument2"].to_dict()
```

```python
bws_testset1 = {}
for i in range(2550,len(testframe)+2550):
    if testframe["argument1"][i] not in bws_testset1.values():
        bws_testset1[i] = testframe["argument1"][i]
    if testframe["argument2"][i] not in bws_testset1.values():
        bws_testset1[i+len(testframe)] = testframe["argument2"][i]
```

```python
bws_testset1
```

```python
f = open("bws_testset1_sentences.txt","w")
f.write(str(bws_testset1))
f.close()
```

```python
similarfilter=bws["score"]>=0.8
notsimilarfilter = bws["score"]<0.8
bwsp = testframe[similarfilter]
bwsf = testframe[notsimilarfilter]
```

```python
bwsf
```

```python
bws_dict = {value:key for key, value in bws_testset1.items()}
```

```python
pairs = {}
for i in range(len(bwsp)):
    pairs.setdefault(bws_dict[bwsp["argument1"].iloc[i]],[]).append(bws_dict[bwsp["argument2"].iloc[i]])
    #pairs.setdefault(test1["#1ID"].iloc[i],[]).append(msrp_test["#2ID"].iloc[i])
```

```python
pairs
```

```python
f = open("bws_testset2_pairs.txt","w")
f.write(str(pairs))
f.close()
```

```python
nopairs = []
for j in range(len(bwsf)):
    nopairs.append(bws_dict[bwsf["argument1"].iloc[j]])
    nopairs.append(bws_dict[bwsf["argument2"].iloc[j]])
nopairs = list(set(nopairs))
```

```python
for s in reversed(range(len(nopairs))):
    if(nopairs[s] in pairs.values() or nopairs[s] in pairs.keys()):
        del nopairs[s]
```

```python
s = open("nopairs_testset2.txt","w")
s.write(str(nopairs))
s.close()
```
