```python
ukp_dict_testset1 = {}
for i in range(2704,len(testframe)+2704):
    if testframe["sentence_1"][i] not in ukp_dict_testset1.values():
        ukp_dict_testset1[i] = testframe["sentence_1"][i]
    if testframe["sentence_2"][i] not in ukp_dict_testset1.values():
        ukp_dict_testset1[i+len(testframe)] = testframe["sentence_2"][i]
```

```python
f = open("ukp_testset4_sentences.txt","w")
f.write(str(ukp_dict_testset1))
f.close()
```

```python
filter1 = testframe["label"] == "HS"
filter2 = testframe["label"] != "HS"
test1 = testframe[filter1]
test2 = testframe[filter2]
```

```python
ukp_dict = {value:key for key, value in ukp_dict_testset1.items()}
```

```python
pairs = {}
for i in range(len(test1)):
    pairs.setdefault(ukp_dict[test1["sentence_1"].iloc[i]],[]).append(ukp_dict[test1["sentence_2"].iloc[i]])
    #pairs.setdefault(test1["#1ID"].iloc[i],[]).append(msrp_test["#2ID"].iloc[i])
```

```python
pairs
```

```python
d = open("ukp_testset4_pairs.txt","w")
d.write(str(pairs))
d.close()
```

```python
nopairs = []
for j in range(len(test2)):
    nopairs.append(ukp_dict[test2["sentence_1"].iloc[j]])
    nopairs.append(ukp_dict[test2["sentence_2"].iloc[j]])
nopairs = list(set(nopairs))
```

```python
for s in reversed(range(len(nopairs))):
    if(nopairs[s] in pairs.values() or nopairs[s] in pairs.keys()):
        del nopairs[s]
```

```python
s = open("nopairs_testset4.txt","w")
s.write(str(nopairs))
s.close()
```
