```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
GM = pd.read_csv("ArgPairs_GM.csv",delimiter=",",encoding='latin1')
DP = pd.read_csv("ArgPairs_DP.csv",delimiter=",",encoding='latin1')
GC = pd.read_csv("ArgPairs_GC.csv",delimiter=",",encoding='latin1')
```

```python
DP
```

```python
GM_sentences = {}
for i in range(len(GM)):
    if GM["sentence_1"][i] not in GM_sentences.values():
        GM_sentences[i] = GM["sentence_1"][i]
    if GM["sentence_2"][i] not in GM_sentences.values():
        GM_sentences[i+len(GM)] = GM["sentence_2"][i]
```

```python
GC_sentences = {}
for i in range(len(GC)):
    if GC["sentence_1"][i] not in GC_sentences.values():
        GC_sentences[i] = GC["sentence_1"][i]
    if GC["sentence_2"][i] not in GC_sentences.values():
        GC_sentences[i+len(GC)] = GC["sentence_2"][i]
```

```python
DP_sentences = {}
for i in range(len(GC)):
    if DP["sentence_1"][i] not in DP_sentences.values():
        DP_sentences[i] = DP["sentence_1"][i]
    if DP["sentence_2"][i] not in DP_sentences.values():
        DP_sentences[i+len(DP)] = DP["sentence_2"][i]
```

```python
f = open("afs_GC_sentences.txt","w")
f.write(str(GC_sentences))
f.close()
```

```python
f = open("afs_GM_sentences.txt","w")
f.write(str(GM_sentences))
f.close()
```

```python
f = open("afs_DP_sentences.txt","w")
f.write(str(DP_sentences))
f.close()
```

```python
filter1 = DP["regression_label"] >= 4
filter2 = DP["regression_label"] <= 4
test1 = DP[filter1]
test2 = DP[filter2]
```

```python
afs_dict = {value:key for key, value in DP_sentences.items()}
```

```python
pairs = {}
for i in range(len(test1)):
    pairs.setdefault(afs_dict[test1["sentence_1"].iloc[i]],[]).append(afs_dict[test1["sentence_2"].iloc[i]])
```

```python
pairs
```

```python
d = open("afs_pairs_DP.txt","w")
d.write(str(pairs))
d.close()
```

```python
nopairs = []
for j in range(len(test2)):
    nopairs.append(afs_dict[test2["sentence_1"].iloc[j]])
    nopairs.append(afs_dict[test2["sentence_2"].iloc[j]])
nopairs = list(set(nopairs))
```

```python
for s in reversed(range(len(nopairs))):
    if(nopairs[s] in pairs.values() or nopairs[s] in pairs.keys()):
        del nopairs[s]
```

```python
s = open("nopairs_DP.txt","w")
s.write(str(nopairs))
s.close()
```
